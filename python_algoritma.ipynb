```python
#####################################################################################################
############################           Python Algoritma Çözümleri        ############################
#####################################################################################################
#####################                        Bölüm 1: Sorular                   #####################
#####################################################################################################
########                                    Kolay Seviye Sorular                             ########
### 1. Listedeki En Büyük Sayıyı Bulma

def en_buyuk_sayi(liste):
    return max(liste)

print("1. Liste için en büyük sayı:", en_buyuk_sayi([12, 45, 7, 33, 88]))
print("2. Liste için en büyük sayı:", en_buyuk_sayi([-100, -7, -200]))
#####################################################################################################
```

```python
### 2. String'i Ters Çevirme

def ters_cevir(metin):
    return metin[::-1]

kelime = "goncimik"
print("Orijinal:", kelime)
print("Ters:", ters_cevir(kelime))
#####################################################################################################
```

```python
### 3. Faktöriyel Hesaplama
# 1.yol

def faktoriyel(n):
    sonuc = 1
    for i in range(1, n + 1):
        sonuc *= i
    return sonuc
print("7! =", faktoriyel(7))

# 2.yol

#import math
#print(math.factorial(7))
#####################################################################################################
```

```python
### 4. Listedeki Çift Sayıları Filtreleme

def cift_sayilari_filtrele(liste):
    return [x for x in liste if x % 2 == 0]

sayilar = [17, 2, 33, 41, 55, 66, 88, 0]
print("Liste:", sayilar)
print("Çift sayılar:", cift_sayilari_filtrele(sayilar))
#####################################################################################################
```

```python
### 5. İki Listenin Birleşimi ve Tekilleştirilmesi

def birlesim_liste(liste1, liste2):
    return list(set(liste1 + liste2))

a = [1, 5, 77, 34, 32, 51, 47]
b = [1, 77, 47, 99, 0, 3]

print("Birleştirilmiş ve tekilleştirilmiş liste:", birlesim_liste(a, b))
#####################################################################################################
```

```python
#####################################################################################################
#####################                       Orta Seviye Sorular                 #####################
#####################################################################################################

### 6. Palindrom Kontrolü

def palindrom_mu(s):
    t = ''.join(c.lower() for c in s if c.isalnum())
    return t == t[::-1]

print(palindrom_mu("gogog"))
#####################################################################################################
```

```python
### 7. İki Sayının En Büyük Ortak Böleni (EBOB)
# 1.yol

def ebob(a, b):
    while b != 0:
        a, b = b, a % b
    return a

print(ebob(54, 24))
print(ebob(7, 13))

# 2.yol
#import math
#print(math.gcd(54, 24))
#print(math.gcd(7, 13))
#####################################################################################################
```

```python
### 8. Anagram Kontrolü
# 1.

from collections import Counter

def anagram_mi(s1, s2):
    return Counter(s1.lower().replace(" ", "")) == Counter(s2.lower().replace(" ", ""))

print(anagram_mi("Dormitory", "Dirty room"))

#2.
#def anagram_mi(a, b):
#return sorted(a.lower().replace(" ", "")) == sorted(b.lower().replace(" ", ""))
#print(anagram_mi("gogo", "ogog"))
#####################################################################################################
```

```python
### 9. Listedeki Eksik Sayıyı Bulma

def eksik_sayi(liste):
    n = len(liste)
    beklenen_toplam = n * (n + 1) // 2
    gercek_toplam = sum(liste)
    return beklenen_toplam - gercek_toplam

print(eksik_sayi([9, 6, 4, 2, 3, 5, 7, 0, 1]))
#####################################################################################################
```

```python
### 10. Listeyi Belirli Bir Değere Göre Döndürme (Rotate)

def rotate (liste, k):
    k %= len(liste)
    return liste[-k:] + liste[:-k]

print(rotate([1, 2, 3, 4, 5], 2))
  #####################################################################################################
```

```python
#####################################################################################################
#####################                       Zor Seviye Sorular                  #####################
#####################################################################################################
### 11. İki Sıralı Listenin Medyanını Bulma

def iki_sirali_listenin_medyani(a, b):
    c = sorted(a + b)
    n = len(c)
    mid = n // 2
    return (c[mid] if n % 2 else (c[mid - 1] + c[mid]) / 2)

print(iki_sirali_listenin_medyani([1, 3], [2]))
#####################################################################################################
```

```python
### 12. En Uzun Palindromik Alt String

def longest_palindrome(s):
    res = ""
    for i in range(len(s)):
        for a, b in ((i, i), (i, i+1)):
            while a >= 0 and b < len(s) and s[a] == s[b]:
                if b - a + 1 > len(res): res = s[a:b+1]
                a -= 1; b += 1
    return res

print(longest_palindrome("gogog"))
print(longest_palindrome("gogo"))
#####################################################################################################
```

```python
#### 13. Kelime Merdiveni (Word Ladder)

from collections import deque

def ladderLength(begin, end, words):
    if end not in words: return 0
    words = set(words)
    q = deque([(begin, 1)])
    while q:
        w, d = q.popleft()
        if w == end: return d
        for i in range(len(w)):
            for c in 'abcdefghijklmnopqrstuvwxyz':
                nxt = w[:i] + c + w[i+1:]
                if nxt in words:
                    words.remove(nxt)
                    q.append((nxt, d+1))
    return 0

print(ladderLength("hit", "cog", ["hot","dot","dog","lot","log","cog"]))
#####################################################################################################
```

```python
### 14. Maksimum Alt Dizi Toplamı

def max_alt_dizi_toplami(nums):
    en_iyi = anlik = nums[0]
    for x in nums[1:]:
        anlik = max(x, anlik + x)
        en_iyi = max(en_iyi, anlik)
    return en_iyi

print(max_alt_dizi_toplami([5, 4, -1, 7, 8]))
#####################################################################################################
```

```python
### 15. Birleştirme Aralıkları (Merge Intervals)

def birlestir_araliklar(intervals):
    if not intervals:
        return []
    intervals.sort(key=lambda x: x[0])
    birlesik = [intervals[0]]

    for s, e in intervals[1:]:
        son_s, son_e = birlesik[-1]
        if s <= son_e:
            birlesik[-1][1] = max(son_e, e)
        else:
            birlesik.append([s, e])
    return birlesik

print(birlestir_araliklar([[1,4],[4,5]]))
#####################################################################################################
```
