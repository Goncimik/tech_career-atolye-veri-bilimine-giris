## 2. Veri Seti


√áe≈üitli √ºlkelerle ilgili bir dizi **demografik**, **ekonomik** ve **coƒürafi** veriyi i√ßermektedir.  
    
Tabloda:  
- Her bir **satƒ±r** ‚Üí bir √ºlkeyi temsil eder.  
- Her bir **s√ºtun** ‚Üí bu √ºlkelerle ilgili farklƒ± √∂zellikleri g√∂sterir.  

ƒ∞√ßerik:

- **Country**: √úlkenin adƒ±  
- **Region**: √úlkenin bulunduƒüu b√∂lge *(√∂r. Asya, Doƒüu Avrupa)*  
- **Population**: √úlkenin toplam n√ºfusu  
- **Area (sq. mi.)**: √úlkenin y√ºz√∂l√ß√ºm√º *(mil kare olarak)*  
- **Pop. Density (per sq. mi.)**: N√ºfus yoƒüunluƒüu *(mil kare ba≈üƒ±na ki≈üi sayƒ±sƒ±)*  
- **Coastline (coast/area ratio)**: Sahil uzunluƒüunun, √ºlkenin toplam alanƒ±na oranƒ±  
- **Net migration**: Net g√∂√ß oranƒ± *(giren‚Äì√ßƒ±kan n√ºfus farkƒ±)*  
- **Infant mortality (per 1000 births)**: Bebek √∂l√ºm oranƒ± *(1000 doƒüum ba≈üƒ±na)*  
- **GDP ($ per capita)**: Ki≈üi ba≈üƒ±na d√º≈üen Gayri Safi Yurti√ßi Hasƒ±la (GSYƒ∞H)  
- **Literacy (%)**: Okur-yazarlƒ±k oranƒ±  
- **Phones (per 1000)**: Her 1000 ki≈üi ba≈üƒ±na d√º≈üen telefon sayƒ±sƒ±  
- **Arable (%)**: Tarƒ±ma elveri≈üli arazi y√ºzdesi  
- **Crops (%)**: Ekilebilir √ºr√ºnlerin y√ºzdesi  
- **Other (%)**: Diƒüer arazi kullanƒ±mƒ± y√ºzdesi  
- **Climate**: √úlkenin iklim kategorisi *(numerik bir deƒüer)*  
- **Birthrate**: Doƒüum oranƒ±  
- **Deathrate**: √ñl√ºm oranƒ±  
- **Agriculture**: Tarƒ±m sekt√∂r√ºn√ºn ekonomideki payƒ±  
- **Industry**: Sanayi sekt√∂r√ºn√ºn ekonomideki payƒ±  
- **Service**: Hizmet sekt√∂r√ºn√ºn ekonomideki payƒ±

# 3. G√∂revler

## G√∂rev 1: N√ºfusa G√∂re Azalan Sƒ±rada Sƒ±ralama

```python
countries_by_population_desc = df.sort_values(by='Population', ascending=False)
countries_by_population_desc.head(10)
```

## G√∂rev 2: GDP per capita S√ºtununa G√∂re Artan Sƒ±rada Sƒ±ralama


```python
countries_by_gdp_asc = df.sort_values(by='GDP ($ per capita)', ascending=True)
countries_by_gdp_asc.head(10)
```

## G√∂rev 3: Population S√ºtunu 10 Milyonun √úzerinde Olan √úlkeler

```python
pop_over_10m = df[df['Population'] > 10_000_000]
print('N√ºfusu 10 milyonun √ºzerinde olan √ºlke sayƒ±sƒ±:', pop_over_10m.shape[0])
pop_over_10m[['Country', 'Region', 'Population']].head(15)
```

## G√∂rev 4: Literacy (%) S√ºtununa G√∂re Sƒ±ralama ve En Y√ºksek 5 √úlke



```python
top5_literacy = df.sort_values(by='Literacy (%)', ascending=False).head(5)
top5_literacy[['Country', 'Region', 'Literacy (%)']]
```

## G√∂rev 5: Ki≈üi Ba≈üƒ± GSYƒ∞H 10.000'in √úzerinde Olan √úlkeler



```python
gdp_over_10000 = df[df['GDP ($ per capita)'] > 10_000]
print('Ki≈üi ba≈üƒ± GSYƒ∞H 10.000 √ºzerinde olan √ºlke sayƒ±sƒ±:', gdp_over_10000.shape[0])
gdp_over_10000[['Country', 'Region', 'GDP ($ per capita)']].head(15)
```

## G√∂rev 6: En Y√ºksek N√ºfus Yoƒüunluƒüuna Sahip ƒ∞lk 10 √úlke


```python
top10_density = df.sort_values(by='Pop. Density (per sq. mi.)', ascending=False).head(10)
top10_density[['Country', 'Region', 'Pop. Density (per sq. mi.)']]
```
