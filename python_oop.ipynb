```python

##############################################################################################################
##########################                      Python OOP Çözümleri                  ########################
##############################################################################################################
#############################                     Bölüm 1: Sorular                 ###########################
#################################               Kolay Seviye Sorular           ###############################
#### 1. Öğrenci Sınıfı
class Ogrenci:
    def __init__(self, ad, no, notlar=[]):
        self.ad, self.no, self.notlar = ad, no, [n for n in notlar if 0 <= n <= 100]

    def not_ekle(self, n):
        if 0 <= n <= 100: self.notlar.append(n)

    def ortalama(self):
        return round(sum(self.notlar) / len(self.notlar), 2) if self.notlar else 0

    def bilgi(self):
        return f"Öğrenci: {self.ad} ({self.no}), Notlar: {self.notlar}, Ortalama: {self.ortalama()}"


Gonca = Ogrenci("Gonca", "911", [85, 90, 98])
print(Gonca.bilgi())
##############################################################################################################
```

```python
### 2. Hesap Makinesi Sınıfı
class HesapMakinesi:
    def __init__(self): self.bellek = 0
    def topla(self, a, b): return a + b
    def cikar(self, a, b): return a - b
    def carp(self, a, b): return a * b
    def bol(self, a, b): return "Hata: Sıfıra bölme mümkün değil!" if b == 0 else a / b
    def bellek_ekle(self, x): self.bellek += x
    def bellek_sifirla(self): self.bellek = 0

h = HesapMakinesi()
print(h.topla(15, 2))
print(h.bol(10, 0))
##############################################################################################################
```

```python
### 3. Kitap Sınıfı
class Kitap:
    def __init__(self, baslik, yazar, sayfa):
        self.baslik, self.yazar = baslik, yazar
        self.sayfa = max(0, sayfa)
        self.okunan = 0

    def oku(self, s):
        if s <= 0:
            print("Okunan sayfa sayısı pozitif olmalı.")
            return
        self.okunan = min(self.sayfa, self.okunan + s)

    def atla(self, s):
        self.oku(s)

    def bilgi(self):
        durum = (
            "Kitap henüz hiç okunmadı." if self.okunan == 0 else
            "Kitap tamamen bitti!" if self.okunan == self.sayfa else
            f"{self.okunan} / {self.sayfa} sayfa okundu."
        )
        return f"{self.baslik} - {self.yazar} ({self.sayfa} sayfa)\nDurum: {durum}"

k = Kitap("AI", "Gonca", 312)
print(k.bilgi())
k.oku(120)
print(k.bilgi())
k.oku(192)
print(k.bilgi())

##############################################################################################################
```

```python
### 4. Banka Hesabı Sınıfı
class BankaHesabi:
    def __init__(self, no, sahip, bakiye=0):
        self.no, self.sahip, self.bakiye = no, sahip, bakiye
        self.gecmis = []

    def yatir(self, t):
        self.bakiye += t
        self.gecmis.append(f"+{t} TL")

    def cek(self, t):
        if t > self.bakiye:
            return "Hata: Yetersiz bakiye!"
        self.bakiye -= t
        self.gecmis.append(f"-{t} TL")

    def bilgi(self):
        son = self.gecmis[-1] if self.gecmis else "Henüz işlem yok"
        return f"Hesap: {self.no}, {self.sahip} | Bakiye: {self.bakiye} TL, Son işlem: {son}"

h = BankaHesabi("12345", "Gonca", 1000)
h.yatir(500)
h.cek(300)
print(h.bilgi())
##############################################################################################################
```

```python
### 5. Geometrik Şekil Sınıfı
import math

class Sekil:
    def __init__(self):
        pass
    def alan(self): raise NotImplementedError
    def cevre(self): raise NotImplementedError

class Dikdortgen(Sekil):
    def __init__(self, genislik, yukseklik):
        if genislik < 0 or yukseklik < 0: raise ValueError("Negatif boyut olamaz!")
        self.g, self.y = genislik, yukseklik
    def alan(self): return self.g * self.y
    def cevre(self): return 2 * (self.g + self.y)

class Daire(Sekil):
    def __init__(self, yaricap):
        if yaricap < 0: raise ValueError("Negatif yarıçap olamaz!")
        self.r = yaricap
    def alan(self): return round(math.pi * self.r**2, 2)
    def cevre(self): return round(2 * math.pi * self.r, 2)


d = Dikdortgen(5, 3)
print(f"Alan: {d.alan()}, Çevre: {d.cevre()}")
c = Daire(4)
print(f"Alan: {c.alan()}, Çevre: {c.cevre()}")

##############################################################################################################
```

```python
##############################################################################################################
########################                      Orta Seviye Sorular                     ########################
##############################################################################################################
### 6. Çalışan Yönetim Sistemi
class Calisan:
    def __init__(self, ad, maas, departman):
        self.ad, self.maas, self.departman = ad, maas, departman
    def bilgi(self):
        return f"{self.ad} - {self.departman} Departmanı, Maaş: {self.maas} TL"

class Yonetici(Calisan):
    def __init__(self, ad, maas, departman, ekip):
        super().__init__(ad, maas, departman)
        self.ekip = ekip
    def bilgi(self):
        return f"{self.ad} - {self.departman} Departmanı, Yönetici, Ekip: {self.ekip} kişi, Maaş: {self.maas + self.ekip*300} TL"

class Muhendis(Calisan):
    def __init__(self, ad, maas, departman, proje):
        super().__init__(ad, maas, departman)
        self.proje = proje
    def bilgi(self):
        return f"{self.ad} - {self.departman} Departmanı, Mühendis, Proje: {self.proje}, Maaş: {self.maas + self.proje*200} TL"

n = Calisan("AlyaLina", 5000, "IT")
print(n.bilgi())
y = Yonetici("Gonca", 8000, "IT", 5)
print(y.bilgi())
m = Muhendis("Adnan", 7000, "R&D", 3)
print(m.bilgi())

##############################################################################################################
```

```python
### 7. Kütüphane Yönetim Sistemi
class Kitap:
    def __init__(self, ad, yazar):
        self.ad, self.yazar = ad, yazar
        self.odunc = False
        self.kimde = None

class Uye:
    def __init__(self, ad):
        self.ad = ad
        self.kitaplar = []

class Kutuphane:
    def __init__(self):
        self.kitaplar = []

    def kitap_ekle(self, kitap):
        self.kitaplar.append(kitap)

    def odunc_ver(self, kitap, uye):
        if kitap.odunc:
            return f"{kitap.ad} kitabı zaten {kitap.kimde} adlı üyede!"
        if len(uye.kitaplar) >= 3:
            return f"{uye.ad} en fazla 3 kitap alabilir!"
        kitap.odunc, kitap.kimde = True, uye.ad
        uye.kitaplar.append(kitap)
        return f"{kitap.ad} kitabı {uye.ad}'e verildi, Kütüphanede kalan: {self.kalan_sayi()} kitap"

    def geri_al(self, kitap, uye):
        if kitap not in uye.kitaplar:
            return f"{uye.ad} bu kitabı almamış."
        kitap.odunc, kitap.kimde = False, None
        uye.kitaplar.remove(kitap)
        return f"{kitap.ad} kitabı {uye.ad}'den geri alındı."

    def kalan_sayi(self):
        return len([k for k in self.kitaplar if not k.odunc])

k1, k2, k3 = Kitap("AI", "Gonca"), Kitap("ML", "AlyaLina"), Kitap("Python", "Adnan")
u1 = Uye("Ahmet")
kutuphane = Kutuphane()
for k in (k1, k2, k3): kutuphane.kitap_ekle(k)

print(kutuphane.odunc_ver(k1, u1))
print(kutuphane.geri_al(k1, u1))


##############################################################################################################
```

```python
### 8. Ev Otomasyon Sistemi
class Cihaz:
    def __init__(self, ad):
        self.ad = ad
        self.acik = False
    def ac(self): self.acik = True
    def kapa(self): self.acik = False
    def durum(self): return f"{self.ad} {'açık' if self.acik else 'kapalı'}"

class Isik(Cihaz):
    def __init__(self, ad, acik=False):
        super().__init__(ad)
        self.acik = acik
    def durum(self): return f"{self.ad} ışığı {'açık' if self.acik else 'kapalı'}"

class Termostat(Cihaz):
    def __init__(self, ad, sicaklik=20):
        super().__init__(ad)
        self.sicaklik = sicaklik
    def durum(self): return f"{self.ad} ısısı {self.sicaklik}°C"

class GuvenlikKamerasi(Cihaz):
    def __init__(self, ad, aktif=False):
        super().__init__(ad)
        self.acik = aktif
    def durum(self): return f"{self.ad} kamerası {'aktif' if self.acik else 'kapalı'}"

class AkilliEv:
    def __init__(self): self.cihazlar = []
    def ekle(self, cihaz): self.cihazlar.append(cihaz)
    def durumlar(self): return ", ".join(c.durum() for c in self.cihazlar)


ev = AkilliEv()
ev.ekle(Isik("Salon", True))
ev.ekle(Termostat("oturma", 22))
ev.ekle(GuvenlikKamerasi("giriş", True))
print(ev.durumlar())

##############################################################################################################
```

```python
### 9. E-ticaret Sepet Sistemi
class Urun:
    def __init__(self, ad, fiyat, stok): self.ad, self.fiyat, self.stok = ad, fiyat, stok

class Sepet:
    def __init__(self): self.ogeler, self.indirim = [], 0
    def ekle(self, urun, adet):
        if adet <= 0: return "Geçersiz adet!"
        if urun.stok < adet: return "Uyarı: Stok yetersiz!"
        self.ogeler.append((urun, adet)); return "Eklendi"
    def indirim_uygula(self, yuzde): self.indirim = max(0, min(1, yuzde/100))
    def toplam(self): return sum(u.fiyat * a for u,a in self.ogeler)
    def ozet(self):
        t = self.toplam(); ind = round(t * self.indirim, 2); od = round(t - ind, 2)
        return t, ind, od

class Siparis:
    def __init__(self, sepet): self.sepet, self.odendi = sepet, False
    def ode(self):
        for u,a in self.sepet.ogeler: u.stok -= a
        self.odendi = True; return "Ödeme alındı"


laptop = Urun("Laptop", 5000, 10)
sepet = Sepet()
print(sepet.ekle(laptop, 2))
sepet.indirim_uygula(10)
t, ind, od = sepet.ozet()
print(f"Sepet Toplamı: {t} TL, İndirim: {ind} TL, Ödenecek: {od} TL")

laptop2 = Urun("Laptop", 4500, 10)
sepet2 = Sepet(); sepet2.ekle(laptop2, 2); sepet2.indirim_uygula(10)
t2, ind2, od2 = sepet2.ozet()
print(f"Sepet Toplamı: {t2} TL, İndirim: {ind2} TL, Ödenecek: {od2} TL")

##############################################################################################################
```
