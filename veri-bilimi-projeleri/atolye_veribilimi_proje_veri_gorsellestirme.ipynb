-  **Marketing Spend (Pazarlama Harcamasƒ±)**: ≈ûirketin pazarlama ve reklam faaliyetleri i√ßin harcadƒ±ƒüƒ± tutar.  

-  **State (Eyalet)**: ≈ûirketin faaliyet g√∂sterdiƒüi eyalet (√∂rneƒüin, New York, California, Florida).  

-  **Profit (K√¢r)**: ≈ûirketin elde ettiƒüi toplam k√¢r.

```python
# Veri Seti
df = pd.read_csv('50_Startups.csv')
df
```

```python
#  Veri Setini Tanƒ±ma
print("=== Veri Boyutu (Rows, Columns) ===")
print(df.shape)

print("\n=== Veri Tipleri ===")
print(df.dtypes)

print("\n=== Eksik Deƒüer Sayƒ±larƒ± ===")
print(df.isnull().sum())

print("\n=== Temel ƒ∞statistikler ===")
display(df.describe().T)
```

# 3. G√∂revler

## G√∂rev 1: R&D Harcamasƒ± ve K√¢r Arasƒ±ndaki ƒ∞li≈üki (Scatter Plot)


```python

plt.figure(figsize=(8,5))
plt.scatter(df["R&D Spend"], df["Profit"], c=df["R&D Spend"], cmap="viridis", alpha=0.7, edgecolor="k")
plt.colorbar(label="Profit ")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Harcamasƒ± & K√¢r Arasƒ±ndaki ƒ∞li≈üki")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## G√∂rev 2: Y√∂netim Harcamalarƒ± ve K√¢r Arasƒ±ndaki ƒ∞li≈üki (Scatter Plot)


```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Administration"], df["Profit"], c=df["Administration"], cmap="plasma", alpha=0.7, edgecolor="k")
plt.colorbar(label="Profit")
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.title("Y√∂netim Harcamasƒ± & K√¢r Arasƒ±ndaki ƒ∞li≈üki")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

## G√∂rev 3: Eyaletlere G√∂re Ortalama K√¢r (Bar Chart)


```python
state_profit = df.groupby("State", as_index=False)["Profit"].mean()
colors = sns.color_palette("coolwarm", len(state_profit))
plt.figure(figsize=(7,5))
plt.bar(state_profit["State"], state_profit["Profit"],color=colors, edgecolor="black")
plt.xlabel("Eyalet")
plt.ylabel("Ortalama K√¢r")
plt.title("Eyaletlere G√∂re Ortalama K√¢r")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

## G√∂rev 4: Harcama T√ºrlerinin Kar≈üƒ±la≈ütƒ±rmasƒ± (Boxplot)


```python
plt.figure(figsize=(10, 5))
sns.boxplot(
    data=df[["R&D Spend", "Administration", "Marketing Spend"]],
    palette="coolwarm",
    orient="h")
plt.xlabel("Harcama Tutarƒ±")
plt.title("Harcama T√ºrlerinin Kar≈üƒ±la≈ütƒ±rmasƒ± (Yatay Boxplot)")
plt.grid(axis="x", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
